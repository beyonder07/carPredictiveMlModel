```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import roc_auc_score,roc_curve,auc
from sklearn.multiclass import OneVsRestClassifier
from sklearn.preprocessing import label_binarize
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.linear_model import ElasticNet
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import LabelEncoder
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import RandomizedSearchCV
import xgboost as xgb
from xgboost import XGBClassifier
```

```python
#categorical Variable

label_encoder = LabelEncoder()

df['Transport'] = df['Transport'].replace(['Car'], 1)
from sklearn.preprocessing import LabelEncoder




df['Gender'] = label_encoder.fit_transform(df['Gender'])
df['license'] = label_encoder.fit_transform(df['license'])
```

```python
X = df[['Age', 'Gender', 'Engineer', 'MBA', 'Work Exp', 'Salary', 'Distance', 'license']]
y = df['Transport']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print(y_train.unique())
print(y_train.dtype)
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')
print(classification_report(y_test, y_pred))
```

```python
cm=confusion_matrix(y_test,y_pred)
print(cm,"\n")
```

```python
plt.figure(figsize=(6,6))
sns.set(font_scale=1.2)
sns.heatmap(cm,annot=True,fmt="d")
```

```python
rfc=RandomForestClassifier(n_estimators=200)
rfc.fit(X_train,y_train)
rfc_pred=rfc.predict(X_test)
```

```python
cm=confusion_matrix(y_test,rfc_pred)
print(cm,"\n")
```

```python
accuracy = accuracy_score(y_test, rfc_pred)

print(f'Accuracy: {accuracy}')
print(classification_report(y_test, rfc_pred))
```

```python
plt.figure(figsize=(6,6))
sns.set(font_scale=1.2)
sns.heatmap(cm,annot=True,fmt="d")
```

```python
svm=SVC(probability=True)
svm.fit(X_train,y_train)
svm_pre=svm.predict(X_test)
```

```python
cm=confusion_matrix(y_test,svm_pre)
print(cm,"\n")
```

```python
accuracy = accuracy_score(y_test, svm_pre)

print(f'Accuracy: {accuracy}')
print(classification_report(y_test, svm_pre))
```

```python
plt.figure(figsize=(6,6))
sns.set(font_scale=1.2)
sns.heatmap(cm,annot=True,fmt="d")
```

```python
param_gridL = {
    'C': [0.001, 0.01, 0.1, 1.0, 10.0],
    'penalty': ['l1', 'l2', 'elasticnet'],
    'solver': ['lbfgs', 'liblinear', 'newton-cg'],
    'max_iter': [100, 200, 300],
}
param_gridR = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
}
param_gridS = {
     'C': [0.1, 1],
    'kernel': ['linear', 'rbf'],
    'gamma': [0.1, 1],
    'tol': [1e-4],
}
```

```python
grid_search = GridSearchCV(estimator=model, param_grid=param_gridL, cv=5,scoring="accuracy")
grid_search.fit(X_train, y_train)
# Print the best hyperparameters
best_lr_model=grid_search.best_estimator_
y_lr_pred=best_lr_model.predict(X_test)
cm=confusion_matrix(y_test,y_lr_pred)
print(confusion_matrix(y_test,y_lr_pred),"\n")
print(classification_report(y_test,y_lr_pred),"\n")
print(accuracy_score(y_test,y_lr_pred));

```
